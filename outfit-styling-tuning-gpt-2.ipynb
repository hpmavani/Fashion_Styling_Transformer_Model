```python
def create_dataset(data_file): 
    iterables = []
    item_prices = []
    item_names = [] 
    item_catids = []

    import json
    with open(data_file) as file: 
        data = json.load(file)
        
    for index, d in enumerate(data): 
        outfit = []
        items = [i for i in d["items"]]
        
        for i_index, i in enumerate(items): 
            iterables.append([index, i_index])
            item_names.append(i["name"])
            item_prices.append(i["price"])
            item_catids.append(i["categoryid"])
    
    multindex = pd.DataFrame(iterables, columns=["outfit_index", "item_index"])
    index = pd.MultiIndex.from_frame(multindex)
    df = pd.DataFrame({"item_name": item_names, "item_price": item_prices, "item_catid": item_catids}, index=index)
    return df
    
```

```python
import re 

def preprocess_df(df): 
    df.replace(["", " ", "...", "Polyvore", ''], np.nan, inplace=True)
    df["item_name"] = df["item_name"].str.lower()
    df["item_name"] = df["item_name"].str.replace("t shirt", "t-shirt")
    df["item_name"] = df["item_name"].str.replace(r"[^A-Za-z\s\-]", "", regex=True)
    df[df["item_name"] == ""]["item_name"] = np.nan
    df = df.dropna()
    df["tokenized_item"] = df["item_name"].apply(lambda x: x.split(" "))
    return df
    
```

```python
def create_input_string(df): 
    num_outfits = df.index.levshape[0]
    outfits = []
    for idx in range(num_outfits): 
        outfit_items = df.loc[idx]["tokenized_item"]
        outfit = ""
        for tokens in outfit_items: 
            for i in tokens: 
                outfit = outfit + i + " "
            outfit = outfit.removesuffix(" ")
            outfit += "[SEP]"
        outfit = outfit.removesuffix("[SEP]")
        outfit += "[OUTFIT_END]"
        outfits.append(outfit)
    return outfits
```

```python
from transformers import GPT2LMHeadModel, AutoTokenizer
import torch
```

```python
# device is set to cuda gpu
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# download GPT2 model, tokenizer
model = GPT2LMHeadModel.from_pretrained("gpt2")
tokenizer = AutoTokenizer.from_pretrained("gpt2")

# move model to gpu
model = model.to(device)

```

```python
# GPT2 doesn't have padding by default, so specify padding token

tokenizer.pad_token = tokenizer.eos_token
model.config.pad_token_id = tokenizer.eos_token_id
```

```python
# 17,316 is a little large for our compute; we need to set up a data loader so we can feed in batches 

from torch.utils.data import DataLoader, Dataset, RandomSampler
import torch

class OutfitDataset(Dataset): 
    def __init__ (self, data, tokenizer): 
        self.data = data
        self.input_ids = []
        self.attn_masks = []
        # pad to max_length means that every sequence is padded to the maximum length sequence in the batch
        # sequences that are too big (> 1024 tokens) might be truncated, our descriptions are short so this likely won't happen
        for outfit in data:
            encodings = tokenizer.encode_plus(outfit,
                                             truncation=True,
                                             padding='max_length',
                                             return_tensors='pt'
                                             )
            self.input_ids.append(torch.squeeze(encodings['input_ids'], 0)) #flatten input_ids to 1D
            self.attn_masks.append(torch.squeeze(encodings['attention_mask'], 0)) #flatten attention masks to 1D

    def __len__(self): 
        return len(self.data)
    def __getitem__(self, idx): 
        return self.input_ids[idx], self.attn_masks[idx]
        
train_dataset = OutfitDataset(train_outfits, tokenizer=tokenizer)
valid_dataset = OutfitDataset(valid_outfits, tokenizer=tokenizer)
print(f"input_ids: {train_dataset[0][0]} attn_masks: {train_dataset[0][1]}")
```

```python
test_dataset = OutfitDataset(test_outfits, tokenizer=tokenizer)
```

```python
from torch.utils.data import RandomSampler

# Dataloaders automatically split the dataset into batches for us
# Batch size of 8 leads to out of memory error, so we will use a batch size of 4

train_dataloader = DataLoader(
    train_dataset,
    sampler = RandomSampler(train_dataset), 
    batch_size = 4
)

valid_dataloader = DataLoader(
    valid_dataset,
    sampler = RandomSampler(valid_dataset), 
    batch_size = 4
)

test_dataloader = DataLoader(
    test_dataset,
    sampler = RandomSampler(test_dataset), 
    batch_size=4
)

train_features, train_labels = next(iter(train_dataloader))
print(f"Feature batch shape: {train_features.size()}")
print(f"Labels batch shape: {train_labels.size()}")

valid_features, valid_labels = next(iter(valid_dataloader))
print(f"Feature batch shape: {valid_features.size()}")
print(f"Labels batch shape: {valid_labels.size()}")
```

```python
from torch.optim import SGD
from torch.nn.utils import clip_grad_norm_
import torch.nn as nn
import matplotlib.pyplot as plt
from transformers import get_scheduler
from torch.amp import autocast, GradScaler
import time


def train_val_loop(model, optimizer, train_dl, valid_dl, train_len, valid_len, learning_rate, epochs, start_epoch, start_step, dropout_rate, weight_decay): 
    # set up dropout 
    model.config.dropout = dropout_rate 

    # move model to device
    model = model.to(device)

    # put model in train mode
    model.train()

    total_steps = epochs * len(train_dl)
    scheduler = get_scheduler(
        "linear", optimizer=optimizer, num_warmup_steps=0, num_training_steps=total_steps
    )

    #Set up mixed precision to speed up training
    scaler = GradScaler("cuda") 
    
    #total_steps = int((train_len/4.0) * epochs)
    #train_steps = range(1, total_steps + 1)
    #train_losses = []
    #train_perps = []
    
    #valid_epochs = range(1, epochs + 1)
    #valid_losses = []
    #valid_perps = []
    
    for epoch in range(start_epoch, epochs): 
        steps = []
        train_losses = []
        train_perps = [] 
        
        start_time = time.time()
        total_train_loss = 0 
        total_train_perplexity = 0.0
        
        for step, batch in enumerate(train_dl):
            
            if epoch == start_epoch and step < start_step: #skip steps for checkpoint
                continue
                
            b_input_ids = batch[0].to(device)
            b_labels = batch[0].to(device)
            b_attention = batch[1].to(device)
            
            #forward pass
            with autocast(device_type="cuda"): 
                outputs = model(input_ids=b_input_ids, labels=b_labels, attention_mask=b_attention) #the model will shift labels to the left for next token prediction
                
                loss = outputs.loss 
                perplexity = torch.exp(loss)
                
                train_losses.append(loss.item())
                train_perps.append(perplexity.item())
                
                total_train_loss += loss.item()
                total_train_perplexity += perplexity.item()
                
                print(f"Epoch {epoch + 1}: Step {step}: Loss: {loss}, Perplexity: {perplexity}")
            
            #back_prop
            optimizer.zero_grad()
            
            scaler.scale(loss).backward() #compute gradients

            scaler.unscale_(optimizer)

            #gradient clipping
            clip_grad_norm_(model.parameters(), max_norm=1.0) 

            #update weights 
            scaler.step(optimizer) 
            scaler.update()

            scheduler.step() 
            steps.append(step)
            
            if(step % 1000 == 0 and step != 0): #save checkpoint every 1000 steps 
                torch.save({
                "epoch": epoch,
                "start_epoch": epoch + 1,
                "step": step,
                "model": model.state_dict(),
                "optimizer": optimizer.state_dict(),
                },f"./checkpoints/checkpoint-e{epoch + 1}-{step}.pt")

            #Check on validation every 3000 steps
            if(step % 3000 == 0 and step != 0):
                model.eval()
                total_valid_loss = 0
                total_valid_perp = 0
                with torch.no_grad(): 
                    for val_step, batch in enumerate(valid_dl): 
                        inputs = batch[0].to(device)
                        labels = batch[0].to(device)
                        attention = batch[1].to(device)
                        outputs = model(inputs, labels=labels, attention_mask=attention)
                        total_valid_loss += outputs.loss.item()
                        total_valid_perp += torch.exp(outputs.loss).item()
                    avg_valid_loss = total_valid_loss/len(valid_dl)
                    avg_valid_perp = total_valid_perp/len(valid_dl)
                    print(f"Epoch {epoch + 1}: Average valid loss: {avg_valid_loss: .4f}, Average valid perp: {avg_valid_perp : .4f}")

        # Per Epoch 
        
        avg_train_loss = total_train_loss/len(train_dl)
        avg_train_perp = total_train_perplexity/len(train_dl)

        #Model Convergence Check after epoch 1 is finished
    
        if(epoch == 0): 
            print(f"Epoch {epoch + 1}: Average training loss: {avg_train_loss:.4f}")
            plt.plot(steps, train_losses)
            plt.xlabel("Steps")
            plt.ylabel("Losses")
            plt.title("Train Loss")
            plt.show()
            plt.savefig(f"train_converge_loss_epoch_{epoch + 1}.png")
    
            print(f"Epoch {epoch + 1}: Average train perplexity: {avg_train_perp:.4f}")
            plt.plot(steps, train_perps)
            plt.xlabel("Steps")
            plt.ylabel("Perplexity")
            plt.title("Train Perplexity")
            plt.savefig(f"train_converge_perp_epoch_{epoch + 1}.png")
            plt.show()

        model.eval()
        total_valid_loss = 0
        total_perp_val = 0
        with torch.no_grad(): 
            for step, batch in enumerate(valid_dl): 
                inputs = batch[0].to(device)
                labels = batch[0].to(device)
                attention = batch[1].to(device)
                
                outputs = model(inputs, labels=labels, attention_mask=attention)
                total_valid_loss += outputs.loss.item()
                total_perp_val += torch.exp(outputs.loss).item()
                
        avg_valid_loss = total_valid_loss/len(valid_dl)
        avg_val_perp = total_perp_val/len(valid_dl)
        overfit_gap = avg_valid_loss - avg_train_loss
        
        end_time = time.time() 
        epoch_duration = end_time - start_time

        valid_loss_epochs.append(avg_valid_loss)
        valid_perp_epochs.append(avg_val_perp)
        
        print(f"Epoch {epoch + 1}: Train Loss = {avg_train_loss: .4f}, Valid loss: {avg_valid_loss: .4f}, Gap = {overfit_gap: .4f}, Train Perp: {avg_train_perp: .4f}, Valid Perp: {avg_val_perp : .4f}")
        print(f"Duration: {epoch_duration}")

        # save model state at the end of every epoch
        torch.save({
        "epoch": epoch,
        "start_epoch": epoch + 1,
        "model": model.state_dict(),
        "optimizer": optimizer.state_dict(),
        }, f"./checkpoints/epochv2_{epoch + 1}.pt")
```

```python
import gc


def load_checkpoint(epoch=0, step=0, input_file=None): 
    gc.collect()
    torch.cuda.empty_cache()
    checkpoint = None
    
    if input_file == None:
        checkpoint_file = f"./checkpoints/checkpoint-e{epoch}-{step}.pt"
        try: 
            checkpoint = torch.load(checkpoint_file)
        except(e): 
            print(f"Could not load checkpoint epoch {epoch} step {step}", e)
    else: 
        checkpoint = torch.load(input_file)

    model.load_state_dict(checkpoint["model"])
    optim.load_state_dict(checkpoint["optimizer"])
    start_step = checkpoint.get("step", 0)
    start_epoch = checkpoint["start_epoch"]

    model.to(device)
    print(f"Loaded model with start epoch {start_epoch} and start step {start_step}.")
    return start_step, start_epoch
    
```

```python
def test_inference(model, inf=None): 
    model.eval()
    tokenizer.pad_token = tokenizer.eos_token
    input_text = None
    
    if inf == None:
        input_text = "dolce gabbana silk shirt[SEP]long sleeved button detail blazer[SEP]asos temple shoe boots[SEP]"
    else: 
        input_text = inf
    # colour block zip top clutch[OUTFIT_END]
    

    inputs = tokenizer(
        input_text, 
        return_tensors="pt", 
        padding=True,
        truncation=True
    ).to(device)

    
    
    with torch.no_grad():
        output_ids = model.generate(
            input_ids=inputs["input_ids"],
            attention_mask=inputs["attention_mask"],
            max_length=1024,      # Change depending on expected description length
            num_return_sequences=10,
            do_sample=True,     # Enables sampling (creative generation)
            top_k=50,           # Limits sampling pool
            top_p=0.95,         # Nucleus sampling
            temperature=0.9,    # Controls randomness
            pad_token_id=tokenizer.pad_token_id,
            repetition_penalty=1.2
        )

    generated_texts = [tokenizer.decode(g, skip_special_tokens=False) for g in output_ids]

    for g in generated_texts: 
        end_index = g.find('[OUTFIT_END]') + len('[OUTFIT_END]')
        if end_index  != -1: 
            print(g[len(input_text):end_index])
```
